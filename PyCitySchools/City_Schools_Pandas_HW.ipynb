```python
import pandas as pd
import numpy as np
```

```python
#read the csv
school_info = "Resources/schools_complete.csv"
student_info = "Resources/students_complete.csv"

#convert to dataframe
sch_data = pd.read_csv(school_info)
stu_data = pd.read_csv(student_info)

#merge the dataframes
sch_com = pd.merge(stu_data, sch_data, how="left", on=["school_name", "school_name"])
#sch_com.head()
sch_com.tail()
```

**District Summary**

```python
#number of schools - pandas has unique
total_sch = len(sch_com["school_name"].unique())
total_sch
```

```python
#number of students, student name gives wrong total
total_stu = len(sch_com["Student ID"].unique())
total_stu
```

```python
#total budget - pandas has functions fot he columns, 
#forgot to call budget from just scools and not all
total_bud = sch_data["budget"].sum()
total_bud
```

```python
#Average Math Score
total_math_avg = sch_com["math_score"].mean()
total_math_avg
```

```python
#Average reading Score
total_read_avg = sch_com["reading_score"].mean()
total_read_avg
```

```python
# %passing Math - add pass fail columns into the df 
# best using numpy function where, than pandas where function
sch_com["Pass/Fail Math"] = np.where(stu_data["math_score"] >= 70,"pass","fail")
sch_com.head()
```

```python
#now get the % of passing students - pass and fail totals match # of students
num_math_pass = (sch_com["Pass/Fail Math"].values == "pass").sum()
#perc_math_pass cheked number of pass
perc_math_pass = (num_math_pass/total_stu)*100
perc_math_pass
```

```python
# Passing Reading - same approach as Math
sch_com["Pass/Fail Reading"] = np.where(stu_data["reading_score"] >= 70,"pass","fail")
sch_com.head()
```

```python
num_reading_pass = (sch_com["Pass/Fail Reading"].values == "pass").sum()
#perc_math_pass cheked number of pass
perc_reading_pass = (num_reading_pass/total_stu)*100
perc_reading_pass
```

```python
#overall pass percent of both math and reading
over_pass_per = (perc_math_pass + perc_reading_pass) /2
over_pass_per
```

```python
district_summary = [{"Total Schools":total_sch,
                     "Total Students":total_stu,
                     "Total Budget":total_bud,
                     "Average Math Score":total_math_avg,
                     "Average Reading Score":total_read_avg,
                     "% Passing Math":perc_math_pass,
                     "% Passing Reading":perc_reading_pass,
                     "% Overall Passing":over_pass_per
}]
dis_table = pd.DataFrame(district_summary)
dis_table
```

```python
#found that true/false boolean does appear in mean function
#need this for the grouping convention
sch_com["pass math"]= sch_com["math_score"] >=70
sch_com["pass reading"]= sch_com["reading_score"] >=70
sch_com.head()
```

**School Summary**

```python
# have to group things by school name, 
#using mean, will provide the average of all the data we need
sch_group = sch_com.groupby(["school_name"]).mean()
#sch_group
sch_group["Per Student Budget"] = sch_group["budget"]/sch_group["size"]
#sch_check = sch_group["size"].sum()
#sch_check
sch_group["% Passing Math"] = sch_group["pass math"]*100
sch_group["% Passing Reading"] = sch_group["pass reading"]*100
sch_group["% Overall Passing"] = (sch_group["% Passing Math"] + 
                                  sch_group["% Passing Reading"])/2
#sch_group
# I can merge only the specific column I want by specifying
# use school_name as key and type is the column I want from data
sch_summary = sch_group.merge(sch_data[["school_name","type"]], how="left", on=["school_name"])
sch_summary.head()
```

```python
sch_summary_df = pd.DataFrame({"School Name":sch_summary["school_name"],
                               "School Type":sch_summary["type"],
                               "Total Students":sch_summary["size"],
                               "Total School Budget":sch_summary["budget"],
                               "Per Student Budget":sch_summary["Per Student Budget"],
                               "Average Math Score":sch_summary["math_score"],
                               "Average Reading Score":sch_summary["reading_score"],
                               "% Passing Math":sch_summary["% Passing Math"],
                               "% Passing Reading":sch_summary["% Passing Reading"],
                               "% Overall Passing":sch_summary["% Overall Passing"]
                               })
sch_summary_df
```

**Top Performing Schools (By % Overall Passing)**

```python
tfs = sch_summary_df.sort_values(["% Overall Passing"], ascending = False)
tfs.head()
```

**Bottom Performing Schools (By % Overall Passing)**

```python
bfs = sch_summary_df.sort_values(["% Overall Passing"], ascending = True)
bfs.head()
```
